```python
import pandas as pd
import numpy as np

# Generate simple sample sales data
np.random.seed(42)

dates = pd.date_range(start="2024-01-01", end="2024-03-31")
products = ["Shirt", "Shoes", "Jeans", "Hat"]
stores = ["Store A", "Store B"]

data = []

for date in dates:
    for store in stores:
        for product in products:
            quantity = np.random.randint(1, 20)
            price = np.random.randint(200, 1500)
            revenue = quantity * price
            data.append([date, store, product, quantity, price, revenue])

df = pd.DataFrame(data, columns=["Date", "Store", "Product", "Quantity", "Price", "Revenue"])

df.to_csv("sales_data.csv", index=False)

df.head()
```

```python
import pandas as pd

df = pd.read_csv("sales_data.csv")
df.head()
```

```python
# Total revenue
total_revenue = df["Revenue"].sum()

# Total quantity sold
total_quantity = df["Quantity"].sum()

total_revenue, total_quantity
```

```python
product_revenue = df.groupby("Product")["Revenue"].sum().sort_values(ascending=False)
product_revenue
```

```python
store_revenue = df.groupby("Store")["Revenue"].sum()
store_revenue
```

```python
import matplotlib.pyplot as plt

product_revenue.plot(kind="bar", figsize=(8,4))
plt.title("Revenue by Product")
plt.ylabel("Revenue")
plt.xlabel("Product")
plt.show()
```

```python
df_daily = df.groupby("Date")["Revenue"].sum()

df_daily.plot(figsize=(12,5))
plt.title("Daily Revenue Trend")
plt.ylabel("Revenue")
plt.xlabel("Date")
plt.show()
```
